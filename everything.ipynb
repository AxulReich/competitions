# Importing the data

```python
train = pd.read_csv('data/train.csv')
X = train.drop('CLASS', axis = 1)
y = train.CLASS

test = pd.read_csv('data/test.csv')
X_test = test.drop('Id', axis = 1)
index_test = test.Id
```

```python
print(X.shape)
print(X_test.shape)
```

```python
df = pd.concat([X, X_test], axis = 0)
```

```python
print(df.shape)
print(df.columns)
```

```python
y.value_counts()
```

## Categorial and numeric features

```python
print(df.dtypes)

cat_variables = [i for i in df.columns if df[i].dtype == 'O']
print('-----categorical features:\n%s' % cat_variables)

num_variables = [i for i in df.columns if df[i].dtype != 'O']
print('\n-----numeric features:\n%s' % num_variables)
```

### One-hot encoding

```python
for c in cat_variables:
    temp = pd.get_dummies(df[c], dummy_na=True, prefix=c)
    df = pd.concat([df, temp], axis = 1)
    df.drop(c, axis=1, inplace=True)

print(df.shape)

df.head()
```

```python
for c in df.columns:
    print(c)
```

```python
df.columns = [re.sub('<', '-', c) for c in df.columns]
```
