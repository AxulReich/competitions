## Split the dataset by `var3` and look at the correlations

```python
df.var3.value_counts()
```

```python
df_372 = df[df.var3 == 372]
df_272 = df[df.var3 == 272]
```

```python
# df_372.corr()
```

```python
# df_272.corr()
```

```python
round(df_272.corr() - df_372.corr(), 3)
```

Видим, что в разных группах корреляции между `crm` и `var11`, `crm` и `anc_veh` различаются на `0.196`, а между `crm` и `var15` — на `-0.337`.

#### `crm` и `var11`

```python
print(np.corrcoef(df.crm, df.var11)[0, 1])
print(np.corrcoef(df_272.crm, df_272.var11)[0, 1])
print(np.corrcoef(df_372.crm, df_372.var11)[0, 1])
```

```python
plt.figure(figsize = (8, 4))
plt.scatter(df_272.crm, df_272.var11, c='g');
```

```python
plt.figure(figsize = (8, 4))
plt.scatter(df_372.crm, df_372.var11, c='b');
```

#### `crm` и `anc_veh`

```python
print(np.corrcoef(df.crm, df.anc_veh)[0, 1])
print(np.corrcoef(df_272.crm, df_272.anc_veh)[0, 1])
print(np.corrcoef(df_372.crm, df_372.anc_veh)[0, 1])
```

```python
plt.figure(figsize = (8, 4))
plt.scatter(df_272.crm, df_272.anc_veh, c='g');
```

```python
plt.figure(figsize = (8, 4))
plt.scatter(df_372.crm, df_372.anc_veh, c='b');
```

#### `crm` и `var15`

```python
print(np.corrcoef(df.crm, df.var15)[0, 1])
print(np.corrcoef(df_272.crm, df_272.var15)[0, 1])
print(np.corrcoef(df_372.crm, df_372.var15)[0, 1])
```
